The dataset we are using consists of issued loan data, with fields like interest rate, education, employment title, annual income, a ‘loan grade’, and the status of a loan (ranging from paid off, ongoing, to late). 
Link: https://www.kaggle.com/wendykan/lending-club-loan-data

The problem: The issue of deciding whether to grant particular loan applications or not is very lucrative, to ensure that lenders get their money back. We would like to use our dataset to be able to decide whether or not to issue future loans. We can use the factors listed above to do so. To simplify things, we just want to classify if someone will default on a loan or not - hence, we will be implementing binary classifiers. A potential issue is that the status of a high percentage of the loans is ‘Current’, meaning that we cannot use them to train our model. Regardless, we still have a lot of data points.

First we need to clean up the dataset. We import our data in and drop the records that we don't need. The .csv file that we are using has over 2 million rows, so it does take about a minute to read in the csv.

```python
# import necessary packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree
from sklearn import datasets
import random
from sklearn.ensemble import RandomForestClassifier
```

```python
messy_df=pd.read_csv('loan.csv')
for i in messy_df.columns:
    print(i)
```

Now, we narrow down the number of columns in the dataframe according to our hypothesis. We also add a binary column, 'loan_paid', derived from 'loan_status', to determine if a loan was 'good' or 'bad'. Similarly, we also replace the 'verification_status' column with a binary indicator.
We also check to see what percentage of a column consists of null values, and if the percentage is more than 70%, we drop the column. We can see that 'mths_since_last_delinq' had too many null values, so it got dropped.
For the remaining columns, we fill in the null values with the most common value.

```python
from sklearn.preprocessing import LabelEncoder
df = messy_df.filter(items=['loan_amnt','term', 'int_rate', 'installment', 'emp_title','emp_length', 'home_ownership', 'annual_inc', 'verification_status', 'tot_hi_cred_lim', 'delinq_2yrs','addr_state','mths_since_last_delinq','last_credit_pull_d','loan_status'])
df = df.loc[(messy_df['loan_status']!='Current')]
df['verified'] = 1
df.loc[(df['verification_status']=='Not Verified'), 'verified']= 0

df = df.dropna(axis=1, thresh=int(0.80*len(df)))
df.fillna(method='ffill', inplace=True)

df['last_credit_pull'] = [int(x[-4:]) for x in df['last_credit_pull_d']]
df['term'] = [int(x[:3]) for x in df['term']]

df['loan_paid'] = 1
df.loc[(df['loan_status']!='Fully Paid'), 'loan_paid']= 0

df = df.drop(columns=['verification_status', 'loan_status', 'last_credit_pull_d'])
#converting categorical string data to numerical data
enc = LabelEncoder()
subset = df.select_dtypes(exclude=[np.number])
for i in subset:
    df[i] = enc.fit_transform(df[i])
df.tail(10)
```

Now, we'd like to see if any of the factors that we've selected are highly correlated to each other. In that case, we'd like to keep only one of the factors to prevent skewing the model. To see the correlations between all the features, we can make a heatmap. As we can see below, installment and loan amount are highly correlated, so it doesn't make sense to use both. Also, employment length, last credit pull, and 2 year delinquency shows a very weak correlation with 'loan_paid', so we can drop those columns.

```python
import seaborn as sns
print(df.shape)
plt.figure(figsize=(15,15))
#corr = df.corr()
corr = df.assign(emp_title=df.emp_title.astype('category').cat.codes,
                 emp_length=df.emp_length.astype('category').cat.codes,
                 home_ownership=df.home_ownership.astype('category').cat.codes).corr()
ax = sns.heatmap(
    corr,  center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True,annot=True,
    xticklabels=1, yticklabels=1
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```

Our hypothesis is that we can predict which loan applications are worth approving, using the factors that we chose to keep in df. 

```python
df = df.drop(columns=['installment', 'emp_length', 'addr_state','delinq_2yrs', 'last_credit_pull'])
X=df.drop(['loan_paid'], axis=1)
Y=df['loan_paid']
```

Before implementing the models, we define a function that can take in confusion matrix values and plot a stacked bar chart to show us how good our models are. Ideally we would like to optimise for specificity, that is, lower the number of false positives and increase the true negatives. This is because from a bank's perspective, it would be a worse outcome to make a loss on a loan, as opposed to not approving a loan that could have been paid back.

```python
def plot_mat (true_negative, false_positive, false_negative, true_positive):
    percent_true_negative = ((true_negative)/(false_positive+true_negative))
    percent_false_negative = ((false_negative)/(false_negative+true_positive))
    percent_true_positive = ((true_positive)/(false_negative+true_positive))
    percent_false_positive = ((false_positive)/(false_positive+true_negative))
    true = (percent_true_negative,percent_true_positive)
    false = (percent_false_positive, percent_false_negative)
    ind =  np.array([0,1])
    p1 = plt.bar(ind,true)
    p2 = plt.bar(ind,false,bottom=true)
    plt.ylabel('Percentage valid/ invalid')
    plt.title('Confusion Matrix Plot for Loan Prediction')
    plt.xticks(ind, ('Actual negatives','Actual positives'))
    plt.legend((p1[0], p2[0]), ('% Correct Prediction', '% Incorrect Prediction'))
    plt.show()
```

We will use two classification models to test our hypothesis. First, we try to use a KNN classifier. 

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size =0.2, random_state= 42)
model = KNeighborsClassifier(n_neighbors = 3)
model.fit(x_train, y_train)
predictions = model.predict(x_test)
```

To test its accuracy, we get the confusion matrix values. Even though our sensitivity is pretty good, the specificity is quite bad. This can also be seen in the bar chart that we plot using our values.

```python
from sklearn.metrics import confusion_matrix
true_negative, false_positive, false_negative, true_positive = confusion_matrix(y_test,predictions, labels=[False,True]).ravel()

total = true_negative + false_negative + false_positive + true_positive
print("Sensitivity:", true_positive / (true_positive + false_negative))
print("Specificity:", true_negative / (true_negative + false_positive))
print("Accuracy:", (true_positive + true_negative)/ total)
print("Error:", (false_positive + false_negative)/ total)
print("Precision:", true_positive / (true_positive + false_positive))
plot_mat(true_negative, false_positive, false_negative, true_positive)
```

Now, we use a Random Forest classifier for our predictions. Since a random forest consists of many decision trees, we hope that the results will be a better. A drawback is that since we have so many rows in our dataframe, and are using multiple columns as our features, this algorithm can take a while. We used the default n_estimator value, but we assume that with higher values the accuracy will be better. Unfortunately increasing the n_estimators does make the algorithm take a very long time.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state = 42)
rf = RandomForestClassifier(bootstrap = True, max_features = "sqrt")

# Fit on training data
rf.fit(X_train, Y_train)
rf_pred_train = rf.predict(X_train)
rf_pred_test = rf.predict(X_test)
```

Below, we get the confusion matrix values and plot the bar chart that corresponds. We can see that sensitivity remains very high, and specificity has definitely improved. Despite this, it is still quite low.

```python
true_negative, false_positive, false_negative, true_positive = confusion_matrix(Y_test, rf_pred_test, labels=[False,True]).ravel()
total = true_negative + false_negative + false_positive + true_positive
print("Sensitivity:", true_positive / (true_positive + false_negative))
print("Specificity:", true_negative / (true_negative + false_positive))
print("Accuracy:", (true_positive + true_negative)/ total)
print("Error:", (false_positive + false_negative)/ total)
print("Precision:", true_positive / (true_positive + false_positive))

plot_mat(true_negative, false_positive, false_negative, true_positive)
```

We thought that one of the reasons for the low specificity could be that there are too few loan defaulters compared to people who have paid their loans in our dataset. This could be resulting in skewed predictions. Hence, we decided to remove approximately half of the rows where the loan had been paid. An easy way to do this was to just mod one of the numerical label values by 2.
We still had quite a good sample size, but with a more balanced result. We ran the models again and there was a significant increase in specificity, but sensitivity did go down. However, we decided that the following results were better, comparatively, since optimizing to recognize negatives made more sense.
